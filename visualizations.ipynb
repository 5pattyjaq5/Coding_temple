```python
country_counts = dfl['country'].value_counts()

plt.figure(figsize=(8, 6))
x_axis = np.linspace(0,30, num=30)
plt.stem(country_counts.index, country_counts.values)
plt.xlabel(r'Country', fontname="Times New Roman", fontsize = 20)
plt.ylabel(r'Number of Projects', fontname="Times New Roman", fontsize = 20)
plt.title('Number of Projects by Country', fontname="Times New Roman", fontsize = 20)
plt.tight_layout()
plt.grid(True)
plt.show()
```
