### Problem 1:
Check out this Pokemon API https://pokeapi.co/
Use the requests package to connect to this API and get and store data for 5 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny', attack base_state, hp base_stat, defense base_stat

**BONUS POINTS IF YOU CAN POPULATE THE IMAGE!**

```python
import requests
import pandas as pd 
import json
from PIL import Image
from io import BytesIO
from bs4 import BeautifulSoup
# importing url
pk_url = 'https://pokeapi.co/docs/v2/pokemon?limit=5'

response = requests.get(pk_url)
print(response)



# making sure i have the proper connection 
response.ok

response.content

soup = BeautifulSoup(response.content, 'html.parser')

pokemon_name = []
base_exp = []
pk_url = []
attack_base_state = []
hp_bs = []
def_bs = []

for node in soup.find_all('tr'):
    pokemon_name.append(node.find_all(list=True)[1])
    base_exp.append(node.find_all(int=True)[2])
    pk_url.append(node.find_all(string=True)[3])
    attack_base_state.append(node.find_all(string=True)[4])
    hp_bs.append(node.find_all(string=True)[5])
    def_bs.append(node.find_all(string=True)[6])
    

pokemon_name.pop(0)
base_exp.pop(0)
pk_url.pop(0)
attack_base_state.pop(0)
hp_bs.pop(0)
def_bs.pop(0)

pokemon_data = pd.DataFrame.from_dict({
    'pokemon_name' : pokemon_name,
    'base experience' : base_exp,
    'sprite url' : pk_url,
    'attack base_state' : attack_base_state,
    'hp base_stat' : hp_bs,
     'defense base stat': def_bs,
})

pokemon_data

#def get_pokemon_info(name:list, base_exp:str, pokemon_url:str, attack_base_state:str, hp_bs:int, def_bs:str):
 #   ur1= f'https://pokeapi.co/docs/v2/{name}/{base_exp}/{pokemon_url}/{attack_base_state}/{hp_bs}/{def_bs}/#pokemon-section.json'
 #   response = requests.get(ur1)
 #   soup = BeautifulSoup(response.content, 'html.parser')
#print(soup.prettify())





```
