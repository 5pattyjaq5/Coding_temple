##### [a-z] or [A-Z] - any lowercase/uppercase letters from a to z<br/>[^2] - anything that's not 2

##### Integer Ranges

```python
# what this compile here is going to do is it will seqrch for values in a string obj
# where they match beginning w the value 0-4, the next value being bethween 7-9, and the last value being 0-3
pattern_int = re.compile('[0-4][7-9][0-3]')
int_str = '67383 383'
# search the string for all matching numbers:
random_numbers = pattern_int.search(int_str)
print(random_numbers)
span = random_numbers.span()
print(int_str[span[0]: span[1]])
```

##### Character Ranges

```python
# just as we looked thtough int values usingg a range, we can do the same thing w characters in a string ]
# compile a character pattern and create a string
char_pattern = re.compile('[A-Z][a-z]')
my_string = 'Hello There Mr.Anderson'
# using a .finall() object , this searches the string for all the matching characters and  returns a list of the values 
found = char_pattern.findall(my_string)
print(found)
```

##### \d - look for any digit 0-9<br/>\D - look for anything that isnt a digit

```python
date = "Today is the 15th. In 16 days it will be the 1st."


pattern_num = re.compile('\d{1,2}[\D]{2}')
found_date = pattern_num.findall(date)
print(found_date)
```

##### \s - look for any white space<br/>\S - look for anything that isnt whitespace

```python
string = 'Are you    afraid of the dark?'

# something to remember is that whitespace is any form of space within a string'
# it does not need to occur at the begining or end, but again can vbe anywhere 

pattern_no_spaces = re.compile('\S[a-z]+')
found_dark = pattern_no_spaces.findall(string)
print(found_dark)

#pattern with any whitespace
pattern_space = re.compile('\s')
found_space = pattern_space.findall(string)
print(found_space)
```

##### \b - look for boundaries or edges of a word<br/>\B - look for anything that isnt a boundary

```python
# compile that uses a string object and a \b
pattern_bound = re.compile(r'\bTheCodingTemple\b')
pattern_found = pattern_bound.findall('TheCodingTemple')
print(pattern_found)

#anything that isnt a boundary 
pattern_bound_none = re.compile(r'\BTheCodingTemple\B')
pattern_none = pattern_bound_none.findall("TheCodingTemple")
print(pattern_none)
```
